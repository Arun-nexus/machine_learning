```python
from sklearn.cluster import KMeans,DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x,y=make_blobs(n_samples=200,n_features=40,centers=2,cluster_std=0.9,random_state=10)
```

```python
inertia=[]
k=range(1,11,1)
print(k)
for i in k:
    Kmeans=KMeans(n_clusters=i,random_state=0)
    Kmeans.fit(x)
    inertia.append(Kmeans.inertia_)
    print(Kmeans.inertia_)
```

```python
plt.plot(k,inertia)
plt.grid(True)
plt.show()
```

```python
model=KMeans(n_clusters=5,random_state=10)
model.fit(x)
y_pred=model.predict(x)
centers=model.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred,cmap="viridis",label="centroid")
plt.scatter(centers[:,0],centers[:,1],marker="x",s=200,linewidths=2,color="red",label="centroids")
plt.legend()
plt.show()
```
